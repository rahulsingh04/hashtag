```python
#create a simple plot using pyplot:-
#adding title and the label:-

import numpy as np
import matplotlib.pyplot as plt
x=np.array([1,2,5,8,12])
print(x)
y=np.array([3,6,7,12,9])
print(y)
plt.plot(x, y)
plt.xlabel("X-LABEL")
plt.ylabel("Y-LABEL")
plt.grid()
plt.title('A SIMPLE GRAPH:-')

plt.show()
```

```python
#creating a various chart using pyplot:--
import numpy as np
import matplotlib.pyplot as plt
subjects=['Math', 'biology', 'chemistry', 'physics', 'english', 'computer']
student1=[97, 68, 59, 81, 77, 92]
student2=[88, 61, 80, 40, 62, 52]
student3=[54, 62, 77, 54, 71, 98]
index=np.arange(6)
width=0.3
plt.plot(index, student1, width, color='aqua', label=student1)
plt.plot(index+width, student2, width, color='red', label=student2)
plt.plot(index+(width*2), student3, width, color='blue', label=student3)
plt.xlabel("X-LABEL")
plt.ylabel("\nY-LABEL")
plt.title("CREATING A VARIOUS CHART:-")
plt.xticks(index+width/2, subjects)
plt.legend()

plt.grid()

plt.show()
```

```python
#CREATING A PIE CHART USING PYPLOT:---
import matplotlib.pyplot as plt
import numpy as np
firms = ["Firm A", "Firm B", "Firm C", "Firm D", "Firm E"]
market_shares=[10, 40, 30, 5, 15]
Explode=[0, 0, 0, 0.5, 0]
plt.pie(market_shares, explode=Explode, labels=firms, shadow=True, startangle=45)
plt.axis('equal')
plt.legend(title='list of the firm:-')
plt.title("PIE CHARTS:-")
plt.show()
```

```python
#creating an histogram using the pyplot:--
import matplotlib.pyplot as plt
import numpy as np
x=np.random.randn(1000)
#print(x)
plt.title("HISTTOGRAM EXAMPLE:-")
plt.xlabel("RANDOM  DATA")
plt.ylabel('FREQUECY')
plt.hist(x, 10, color='r')
plt.grid()
plt.legend("list:-")

plt.show()
```

```python
#create a 3d plot using matplotlib:--
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib
x=np.linspace(-5, 5, 50)
y=np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)

r=np.sqrt(X**2 + Y**2)
z=np.sin(r)
figure=plt.figure(1, figsize=(12, 4))
subplot3d= plt.subplot(111, projection='3d')
surface= subplot3d.plot_surface(x, y, z, rstride=1, cstride=1, cmap=matplotlib.cm.coolwarm ,linewidth=0.5)
plt.show()
```

```python
#CREATE A 3D GRAPH:-
import matplotlib.pyplot  as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib

x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)
figure = plt.figure(1, figsize = (12, 4))
subplot3d = plt.subplot(111, projection='3d')
surface = subplot3d.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=matplotlib.cm.coolwarm, linewidth=0.1)
plt.show()
```

```python
#CREATE A SCATTER PLOT USING PYPLOT:;
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import rand
for color in ["red", "green", "blue"]:
    n=100
    x, y=rand(2, n)
    scale=500.0*rand(n)**5
    plt.scatter(x, y, s=scale, alpha=0.3, edgecolor='blue')
plt.grid(True)
plt.title("SCATTER PLOT")
plt.xlabel("X AXIS")
plt.ylabel("Y AXIS")
plt.show()
```
